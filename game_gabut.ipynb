```python
import random

# Daftar kata
kata_list = ['python', 'program', 'developer', 'hangman', 'keyboard']
kata_rahasia = random.choice(kata_list)
kata_tertebak = ['_'] * len(kata_rahasia)
tebakan_salah = []
kesempatan = 6

print("=== Game Hangman ===")
print("Tebak kata dengan menebak satu huruf per giliran.")
print("Kamu punya", kesempatan, "kesempatan.\n")

while kesempatan > 0 and '_' in kata_tertebak:
    print("Kata:", ' '.join(kata_tertebak))
    print("Tebakan salah:", ', '.join(tebakan_salah))
    huruf = input("Tebak huruf: ").lower()

    if len(huruf) != 1 or not huruf.isalpha():
        print("Masukkan 1 huruf saja!\n")
        continue

    if huruf in kata_rahasia:
        for i in range(len(kata_rahasia)):
            if kata_rahasia[i] == huruf:
                kata_tertebak[i] = huruf
        print("‚úÖ Benar!\n")
    else:
        if huruf not in tebakan_salah:
            tebakan_salah.append(huruf)
            kesempatan -= 1
            print("‚ùå Salah! Kesempatan tersisa:", kesempatan, "\n")
        else:
            print("Kamu sudah menebak huruf itu!\n")

if '_' not in kata_tertebak:
    print("üéâ Selamat! Kamu berhasil menebak kata:", kata_rahasia)
else:
    print("üò¢ Game Over! Kata yang benar adalah:", kata_rahasia)
```

```python
import random

# Data karakter pemain
player = {
    "nama": "Pahlawan",
    "hp": 100,
    "max_hp": 100,
    "atk": 20,
    "level": 1,
    "exp": 0,
    "exp_next": 50,
    "inventory": ["Potion"],
    "skills": ["Slash"],
}

# Daftar musuh
enemies = [
    {"nama": "Goblin", "hp": 60, "atk": 10, "drop": "Potion", "exp": 30},
    {"nama": "Skeleton", "hp": 80, "atk": 15, "drop": "Elixir", "exp": 40},
    {"nama": "Naga", "hp": 120, "atk": 20, "drop": "Fire Sword", "exp": 70},
]

def show_status():
    print(f"\n‚ù§Ô∏è HP: {player['hp']} / {player['max_hp']} | ‚≠ê LVL: {player['level']} | üéí Inventory: {player['inventory']}")

def level_up():
    while player["exp"] >= player["exp_next"]:
        player["level"] += 1
        player["exp"] -= player["exp_next"]
        player["exp_next"] += 50
        player["max_hp"] += 20
        player["atk"] += 5
        player["skills"].append("Power Slash" if "Power Slash" not in player["skills"] else "Mega Strike")
        print(f"\nüî• LEVEL UP! Sekarang level {player['level']}! Serangan dan HP meningkat.")

def use_item():
    print("\nüéí Inventory:", player["inventory"])
    item = input("Gunakan item apa? ").title()
    if item in player["inventory"]:
        if item == "Potion":
            player["hp"] = min(player["max_hp"], player["hp"] + 30)
            print("üß™ Kamu menggunakan Potion dan menyembuhkan 30 HP.")
        elif item == "Elixir":
            player["hp"] = player["max_hp"]
            print("‚ú® Kamu menggunakan Elixir dan memulihkan HP penuh.")
        elif item == "Fire Sword":
            player["atk"] += 10
            print("üî• Kamu menggunakan Fire Sword! Serangan naik 10 poin.")
        player["inventory"].remove(item)
    else:
        print("‚ùå Item tidak ditemukan!")

def choose_skill():
    print("üí• Skill yang tersedia:", player["skills"])
    skill = input("Gunakan skill apa? ").title()
    if skill in player["skills"]:
        if skill == "Slash":
            return random.randint(player["atk"], player["atk"] + 5)
        elif skill == "Power Slash":
            return random.randint(player["atk"] + 5, player["atk"] + 15)
        elif skill == "Mega Strike":
            return random.randint(player["atk"] + 10, player["atk"] + 20)
    else:
        print("‚ùå Skill tidak tersedia.")
        return 0

def battle(enemy):
    print(f"\n‚öîÔ∏è Kamu bertemu dengan {enemy['nama']}!")
    enemy_hp = enemy["hp"]
    while enemy_hp > 0 and player["hp"] > 0:
        show_status()
        action = input("Aksi (serang / skill / item / kabur): ").lower()

        if action == "serang":
            damage = random.randint(player["atk"] - 5, player["atk"] + 5)
            enemy_hp -= damage
            print(f"üó°Ô∏è Kamu menyerang {enemy['nama']} dan memberi {damage} damage.")
        elif action == "skill":
            damage = choose_skill()
            if damage:
                enemy_hp -= damage
                print(f"üí• Kamu menggunakan skill dan memberi {damage} damage.")
        elif action == "item":
            use_item()
            continue
        elif action == "kabur":
            print("üèÉ‚Äç‚ôÇÔ∏è Kamu melarikan diri dari pertempuran.")
            return
        else:
            print("‚ùì Aksi tidak dikenali.")
            continue

        if enemy_hp > 0:
            enemy_damage = random.randint(enemy["atk"] - 3, enemy["atk"] + 3)
            player["hp"] -= enemy_damage
            print(f"üî• {enemy['nama']} menyerang dan memberi {enemy_damage} damage.")

    if player["hp"] <= 0:
        print("\nüíÄ Kamu kalah...")
    else:
        print(f"\nüéâ Kamu mengalahkan {enemy['nama']}!")
        player["exp"] += enemy["exp"]
        level_up()
        print(f"üéÅ Item drop: {enemy['drop']} | +{enemy['exp']} EXP")
        player["inventory"].append(enemy["drop"])

# Main loop
while player["hp"] > 0:
    input("\nTekan Enter untuk bertemu musuh berikutnya...")
    musuh = random.choice(enemies)
    battle(musuh)

print("\nüèÅ Game Selesai. Terima kasih telah bermain!")
```

```python
import random

# Definisikan kelas untuk karakter dan musuh
class Character:
    def __init__(self, nama, hp, atk, level, exp, inventory, money):
        self.nama = nama
        self.hp = hp
        self.max_hp = hp
        self.atk = atk
        self.level = level
        self.exp = exp
        self.exp_next = 100
        self.inventory = inventory
        self.money = money
        self.skills = ["Basic Attack"]

    def level_up(self):
        while self.exp >= self.exp_next:
            self.level += 1
            self.exp -= self.exp_next
            self.exp_next += 50
            self.max_hp += 20
            self.atk += 5
            print(f"\nüî• {self.nama} mencapai LEVEL {self.level}!")

    def heal(self, amount):
        self.hp = min(self.max_hp, self.hp + amount)

    def take_damage(self, amount):
        self.hp -= amount

    def is_alive(self):
        return self.hp > 0

    def add_skill(self, skill):
        if skill not in self.skills:
            self.skills.append(skill)
            print(f"üß† Kamu belajar skill baru: {skill}!")

    def use_skill(self, skill, enemy):
        if skill == "Fireball":
            damage = random.randint(self.atk + 10, self.atk + 20)
            enemy.take_damage(damage)
            print(f"üî• Kamu menggunakan Fireball dan memberi {damage} damage pada {enemy.nama}!")
        elif skill == "Healing Light":
            heal_amount = 40
            self.heal(heal_amount)
            print(f"‚ú® Kamu menggunakan Healing Light dan sembuh {heal_amount} HP!")
        elif skill == "Thunder Strike":
            damage = random.randint(self.atk + 15, self.atk + 25)
            enemy.take_damage(damage)
            print(f"‚ö° Kamu menggunakan Thunder Strike dan memberi {damage} damage pada {enemy.nama}!")
        else:
            print("‚ùì Skill tidak ditemukan!")

class Enemy:
    def __init__(self, nama, hp, atk, drop, exp, is_boss=False):
        self.nama = nama
        self.hp = hp
        self.atk = atk
        self.drop = drop
        self.exp = exp
        self.is_boss = is_boss

    def take_damage(self, amount):
        self.hp -= amount

    def is_alive(self):
        return self.hp > 0

# Fungsi untuk bertarung
def battle(player, enemy):
    print(f"\n‚öîÔ∏è Kamu bertemu dengan {enemy.nama}! {enemy.nama} HP: {enemy.hp}")
    while player.is_alive() and enemy.is_alive():
        print(f"\n‚ù§Ô∏è HP {player.nama}: {player.hp}/{player.max_hp} | üíÄ HP {enemy.nama}: {enemy.hp}")
        action = input("Pilih aksi (serang / skill / kabur / potion): ").lower()

        if action == "serang":
            damage = random.randint(player.atk - 5, player.atk + 5)
            enemy.take_damage(damage)
            print(f"üó°Ô∏è Kamu menyerang {enemy.nama} dan memberi {damage} damage!")
        elif action == "skill":
            print("üí• Pilih skill yang ingin digunakan:", player.skills)
            skill = input("Pilih skill: ").title()
            player.use_skill(skill, enemy)
        elif action == "kabur":
            print("üèÉ‚Äç‚ôÇÔ∏è Kamu mencoba kabur dari pertempuran!")
            if random.random() < 0.5:
                print("‚ö†Ô∏è Kamu berhasil kabur!")
                return
            else:
                print("‚ùå Kabur gagal! Naga menyerang!")
                damage = random.randint(enemy.atk - 5, enemy.atk + 5)
                player.take_damage(damage)
                print(f"üî• {enemy.nama} menyerang balik dan memberi {damage} damage!")
        elif action == "potion":
            if "Potion" in player.inventory:
                player.inventory.remove("Potion")
                player.heal(30)
                print("üß™ Kamu menggunakan Potion dan menyembuhkan 30 HP!")
            else:
                print("‚ùå Kamu tidak memiliki Potion!")
        else:
            print("‚ùì Aksi tidak valid!")
            continue

        if enemy.is_alive():
            damage = random.randint(enemy.atk - 5, enemy.atk + 5)
            player.take_damage(damage)
            print(f"üî• {enemy.nama} menyerang dan memberi {damage} damage!")

    if player.is_alive():
        print(f"\nüéâ Kamu berhasil mengalahkan {enemy.nama}!")
        player.exp += enemy.exp
        player.level_up()
        player.inventory.append(enemy.drop)
        player.money += 10
        print(f"üéÅ {enemy.drop} jatuh dari {enemy.nama}. Kamu mendapat +{enemy.exp} EXP dan 10 uang!")
        if enemy.is_boss:
            print(f"\nüèÜ Kamu telah mengalahkan Boss {enemy.nama} dan melanjutkan ke level berikutnya!")
    else:
        print(f"\nüíÄ Kamu kalah dalam pertempuran melawan {enemy.nama}... Game Over.")

# Fungsi untuk membeli item dari toko
def shop(player):
    print(f"\nüí∞ Uang kamu: {player.money} gold")
    print("Toko menawarkan:")
    print("1. Potion (50 gold) - +30 HP")
    print("2. Elixir (100 gold) - Full HP")
    print("3. Fireball Skill (200 gold) - Serangan api")
    print("4. Thunder Strike Skill (300 gold) - Serangan petir")
    choice = input("Pilih item untuk dibeli (1/2/3/4): ")

    if choice == "1" and player.money >= 50:
        player.inventory.append("Potion")
        player.money -= 50
        print("üß™ Kamu membeli Potion!")
    elif choice == "2" and player.money >= 100:
        player.inventory.append("Elixir")
        player.money -= 100
        print("‚ú® Kamu membeli Elixir!")
    elif choice == "3" and player.money >= 200:
        player.add_skill("Fireball")
        player.money -= 200
        print("üî• Kamu membeli skill Fireball!")
    elif choice == "4" and player.money >= 300:
        player.add_skill("Thunder Strike")
        player.money -= 300
        print("‚ö° Kamu membeli skill Thunder Strike!")
    else:
        print("‚ùå Uangmu tidak cukup atau pilihan tidak valid!")

# Fungsi utama untuk menjalankan dungeon crawl
def dungeon_crawl():
    player = Character("Pahlawan", 100, 20, 1, 0, ["Potion"], 0)
    print(f"Selamat datang di Dungeon Crawl, {player.nama}!")

    # Daftar musuh dan Boss
    enemies = [
        Enemy("Goblin", 50, 10, "Potion", 30),
        Enemy("Skeleton", 80, 15, "Elixir", 40),
        Enemy("Dragon", 150, 25, "Dragon Sword", 100, is_boss=True)
    ]

    level = 1
    while player.is_alive():
        print(f"\nüîë Kamu memasuki ruangan level {level}...")

        if random.random() < 0.7:
            # Musuh biasa
            enemy = random.choice(enemies[:-1])  # Kecuali boss
            battle(player, enemy)
        else:
            # Musuh Boss di akhir level
            print(f"üî• Boss Fight! Kamu bertemu dengan Boss level {level}!")
            enemy = enemies[-1]  # Boss
            battle(player, enemy)
            level += 1

        # Pilihan melanjutkan atau keluar
        if level > 3:
            print(f"\nüéâ {player.nama} mencapai level {level}! Kamu keluar dari Dungeon dengan kemenangan!")
            break

        # Toko
        shop(player)

    if not player.is_alive():
        print("\nüèÅ Game Over. Coba lagi nanti!")

# Mulai game
dungeon_crawl()
```
